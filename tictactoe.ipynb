```python
from tkinter import *
import tkinter.messagebox
tk = Tk()
tk.title("TIC TAC TOE GAME")

pa = StringVar()
pb = StringVar()
player1 = StringVar()
player2 = StringVar()

player_1_name = Entry(tk, textvariable = player1, bd = 5)
player_1_name.grid(row = 1, column = 1, columnspan = 8)
player_2_name = Entry(tk, textvariable = player2, bd = 5)
player_2_name.grid(row = 2, column = 1, columnspan = 8)

playerclick= True
flag = 0

def disbutton():
    button1.configure(state=DISABLED)
    button2.configure(state=DISABLED)
    button3.configure(state=DISABLED)
    button4.configure(state=DISABLED)
    button5.configure(state=DISABLED)
    button6.configure(state=DISABLED)
    button7.configure(state=DISABLED)
    button8.configure(state=DISABLED)
    button9.configure(state=DISABLED)
   
   
def btnclick(buttons):
    global playerclick, pa, pb, player_1_name, player_2_name, flag
    if buttons["text"] == " " and playerclick == True:
        buttons["text"] = "O"
        playerclick = False
        pb = player2.get() + " Wins the Game!"
        pa = player1.get() + " Wins the Game!"
        CheckVar()
        flag += 1
       
    elif buttons["text"] == " " and playerclick == False:
        buttons["text"] = "X"
        playerclick = True
        CheckVar()
        flag += 1
    else:
        tkinter.messagebox.showinfo("TIC TAC TOE GAME", "Please Choose a Vacant Space!")
       

def CheckVar():
    if (button1["text"] == "O" and button4["text"] == "O" and button7["text"] == "O" or
        button2["text"] == "O" and button5["text"] == "O" and button8["text"] == "O" or
       
        button3["text"] == "O" and button6["text"] == "O" and button9["text"] == "O" or  
        button7["text"] == "O" and button5["text"] == "O" and button3["text"] == "O" or
       
        button1["text"] == "O" and button5["text"] == "O" and button9["text"] == "O" or
        button1["text"] == "O" and button2["text"] == "O" and button3["text"] == "O" or
       
        button4["text"] == "O" and button5["text"] == "O" and button6["text"] == "O" or
        button7["text"] == "O" and button8["text"] == "O" and button9["text"] == "O"):
       
        disbutton()    
        tkinter.messagebox.showinfo("TIC TAC TOE GAME", pa)
   
    elif (flag == 8):
        tkinter.messagebox.showinfo("TIC TAC TOE GAME", "It's tie between both the players.")
       
    elif (button1["text"] == "X" and button4["text"] == "X" and button7["text"] == "X" or
          button2["text"] == "X" and button5["text"] == "X" and button8["text"] == "X" or
         
          button3["text"] == "X" and button6["text"] == "X" and button9["text"] == "X" or
          button7["text"] == "X" and button5["text"] == "X" and button3["text"] == "X" or
         
          button1["text"] == "X" and button5["text"] == "X" and button9["text"] == "X" or
          button1["text"] == "X" and button2["text"] == "X" and button3["text"] == "X" or
         
          button4["text"] == "X" and button5["text"] == "X" and button6["text"] == "X" or
          button7["text"] == "X" and button8["text"] == "X" and button9["text"] == "X"):
       
        disbutton()
        tkinter.messagebox.showinfo("TIC TAC TOE GAME", pb)
       
       
buttons = StringVar()

button1 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button1))
button1.grid(row = 3, column = 1)

button2 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button2))
button2.grid(row = 4, column = 1)

button3 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button3))
button3.grid(row = 5, column = 1)

button4 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button4))
button4.grid(row = 3, column = 2)

button5 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button5))
button5.grid(row = 4, column = 2)

button6 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button6))
button6.grid(row = 5, column = 2)

button7 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button7))
button7.grid(row = 3, column = 3)

button8 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button8))
button8.grid(row = 4, column = 3)

button9 = Button(tk, text = " ", font = 'Times 20 bold', bg = 'yellow', fg = 'black', height = 4, width = 8, command=lambda: btnclick(button9))
button9.grid(row = 5, column = 3)

label = Label (tk, text = "Player 1:- ", font = 'Times 20 bold', bg = 'white', fg ='black', height = 2, width = 8)
label.grid (row = 1, column = 0)

label = Label (tk, text = "Player 2:- ", font = 'Times 20 bold', bg = 'white', fg = 'black', height = 2, width = 8)
label.grid (row = 2, column = 0)            

tk.mainloop()
```
